```python
# importing some libraryies for EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
'''1.Load the dataframe and ensure data quality by checking for missing values and duplicate rows.
 Handle missing values and remove duplicate rows if necessary.'''

# Ans
# Load the dataset
df = pd.read_csv('spotify.csv')  # Replace 'spotify_data.csv' with the actual file name

# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# Check for duplicate rows
duplicate_rows = df.duplicated().sum()
print("\nDuplicate Rows:", duplicate_rows)

# Handle missing values if necessary
# For example, you can drop rows with missing values using df.dropna()
df.dropna(inplace=True)

# Remove duplicate rows if necessary
# For example, you can drop duplicate rows using df.drop_duplicates()
df.drop_duplicates(inplace=True)

# Confirm that missing values and duplicate rows are handled
print("After Handling Missing Values and Duplicates:")
print("Missing Values:\n", df.isnull().sum())
print("Duplicate Rows:", df.duplicated().sum())
```

```python
'''2.What is the distribution of popularity among the tracks in the dataset? Visualize it using a histogram.'''


# Ans
# Check the distribution of 'Popularity' using a histogram
plt.figure(figsize=(10, 6),facecolor='lightgray')
sns.histplot(df['Popularity'], bins=30, kde=True)
# Set the font properties
font = FontProperties()
font.set_family('serif')  # Change font family here
font.set_style('italic')  # Change font style here
font.set_weight('bold')   # Change font weight here

plt.xlabel('Popularity', color='#957BF4', fontproperties=font, fontsize= 18)
plt.ylabel('Frequency', color='#957BF4',fontproperties=font, fontsize= 18)
plt.title('Distribution of Popularity Among Tracks',color='#2FC4A4',  fontproperties=font, fontsize= 18)

plt.grid(True)
plt.show()
```

```python
'''3.Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot'''


# Ans
# Create a scatter plot to explore the relationship between 'Popularity' and 'Duration (ms)'
plt.figure(figsize=(10, 6),facecolor='#A6DDDA')
sns.scatterplot(x='Duration (ms)', y='Popularity', data=df)
# Set the font properties
font = FontProperties()
font.set_family('serif')  # Change font family here
font.set_style('italic')  # Change font style here
font.set_weight('bold')   # Change font weight here
plt.xlabel('Duration (ms)', color='red',  fontproperties=font, fontsize= 16)
plt.ylabel('Popularity', color='red',  fontproperties=font, fontsize= 16)
plt.title('Relationship between Popularity and Duration of Tracks',  color='#3CEE1C',  fontproperties=font, fontsize= 18)
plt.grid(True)
plt.show()
```

```python
'''4.Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist
using a countplot.'''

# Ans

# Count the number of tracks for each artist
artist_track_count = df['Artist'].value_counts()

# Display the count of tracks for each artist using a countplot
plt.figure(figsize=(12, 20),facecolor="lightblue")
sns.countplot(y='Artist', data=df, order=artist_track_count.index)
font = FontProperties()
font.set_family('Monospace')  # Change font family here
font.set_style('italic')  # Change font style here
font.set_weight('bold')   # Change font weight here

plt.xlabel('Number of Tracks',color = '#F8C213',  fontproperties=font, fontsize=16)
plt.ylabel('Artist',color = '#F8C213',  fontproperties=font, fontsize=16)
plt.title('Number of Tracks by Artist', color="red", fontproperties=font, fontsize=18)

plt.show()

# Print the artist with the highest number of tracks
highest_tracks_artist = artist_track_count.idxmax()
print(f"The artist with the highest number of tracks is {highest_tracks_artist} with {artist_track_count.max()} tracks.")
```

```python
'''5.5.What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.'''

#Ans
# Sort the dataframe by 'Popularity' in ascending order
least_popular_tracks = df.sort_values(by='Popularity').head(5)

# Display the artist name and track name for each of the top 5 least popular tracks
for index, row in least_popular_tracks.iterrows():
    artist_name = row['Artist']
    track_name = row['Track Name']
    print(f"Artist: {artist_name}, Track: {track_name}")
```

```python
'''6..Among the top 5 most popular artists, which artist has the highest popularity on average?
 Calculate and display the average popularity for each artist.'''

# Ans
# Calculate the average popularity for each artist
artist_avg_popularity = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False)

# Get the top 5 most popular artists
top_5_popular_artists = artist_avg_popularity.head(5)

# Display the average popularity for each artist
print("Average Popularity for Each Artist:")
print(top_5_popular_artists)

# Identify the artist with the highest average popularity among the top 5 most popular artists
highest_avg_popularity_artist = top_5_popular_artists.idxmax()
highest_avg_popularity = top_5_popular_artists.max()
print(f"\nThe artist with the highest average popularity among the top 5 most popular artists is '{highest_avg_popularity_artist}' with an average popularity of {highest_avg_popularity:.2f}.")
```

```python
'''7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist.'''

# Ans
# Calculate the total popularity for each artist
artist_total_popularity = df.groupby('Artist')['Popularity'].sum().sort_values(ascending=False)

# Get the top 5 most popular artists
top_5_popular_artists = artist_total_popularity.head(5)

# Filter the dataset to include only entries for the top 5 most popular artists
top_5_artists_df = df[df['Artist'].isin(top_5_popular_artists.index)]

# Identify the most popular track for each artist
most_popular_tracks = top_5_artists_df.groupby('Artist').apply(lambda x: x.loc[x['Popularity'].idxmax()])

# Display the most popular track for each artist
for index, row in most_popular_tracks.iterrows():
    artist_name = row['Artist']
    track_name = row['Track Name']
    print(f"Artist: {artist_name}, Most Popular Track: {track_name}")
```

```python
'''8.Visualize relationships between multiple numerical variables simultaneously using a pair plot.'''

# Ans
# Select numerical columns for the pair plot
numerical_columns = ['Popularity', 'Duration (ms)']

# Create a pair plot
sns.pairplot(df[numerical_columns])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)  # Add a title above the plot
plt.show()
```

```python
'''Does the duration of tracks vary significantly across different artists?
Explore this visually using a box plot or violin plot'''

# Ans
# Box Plot
# Create a box plot to visualize the duration of tracks across different artists
plt.figure(figsize=(15,6))
sns.boxplot(x='Artist', y='Duration (ms)', data=df)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.title('Duration of Tracks Across Different Artists (Box Plot)')
plt.grid(True)
plt.show()
```

```python
'''10. How does the distribution of track popularity vary for different artists?
Visualize this using a swarm plot or a violin plot.'''

# Ans
# Swarm Plot:
# Create a swarm plot to visualize the distribution of track popularity across different artists
plt.figure(figsize=(20, 10))
sns.swarmplot(x='Artist', y='Popularity', data=df)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.title('Distribution of Track Popularity Across Different Artists (Swarm Plot)')
plt.grid(True)
```
